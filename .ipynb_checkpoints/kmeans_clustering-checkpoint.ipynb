```python
class MRS_Clusterer:
    """A custom brain clustering algorithm combining K-Means techniques , Autoencoder and PCA"""
    def __init__(self,k):
        self.k = k    #number of clusters
        self.means=None
        self.batch_size = 7
        self.batch_shape =(self.batch_size, 2048,1, 1)
        self.mrs_data_input_shape =2048*28
        self.lr =0.001
        self.num_epochs = 50
        self.n_hidden =64*64
        self.X =tf.placeholder(tf.float32,[None,28])
        self.display_step=1
        self.data=self.load_dataset()


    def encoder(self):
        """Encoder hidden layer with relu activation"""
        layer_h1=tf.nn.relu(tf.add(tf.matmul(self.data, tf.random_normal([self.mrs_data_input_shape,self.n_hidden]),tf.random_normal([self.n_hidden]))))
        return layer_h1
        
        
    def decoder(self,X):
        """Decoder network"""
        layer_h2= tf.nn.relu(tf.add(tf.matmul(X, tf.transpose(tf.random_normal([self.mrs_data_input_shape,self.n_hidden])),tf.random_normal([self.mrs_data_input_shape]))))
        return layer_h2
    

    
    def pca(self):
        """performs principal component analysis using sklearn"""
        
        scaled_inputs = scale(self.data)
        cov_matrix = PCA(n_components =28)
        
        cov_matrix.fit(scaled_inputs)
        
        variance = cov_matrix.explained_variance_ratio_
        var =np.cumsum(np.round(cov_matrix.explained_variance_ratio_,decimals=3)*100)
        
    
        return plt.plot(var)
              
        
    def classify(self):
        """returns the index of the cluster closest to the input"""
        return min(range(self.k),key=lambda i: tf.pow(self.data -self.means[i]))
        
    def train(self):
        """training MRS autoencoder"""
        cost=tf.div(tf.reduce_sum(tf.square(tf.subtract(dataset,
                                                 self.decoder(self.encoder(self.data))))), 
                  tf.constant(float(self.batch_size)))        
        optimizeer =tf.train.GradientDescentOptimizer(lr).minimize(cost)
        init =tf.global_variable_initializer()
        started =time.time()
        with tf.Session() as sess:
            sess.run(init)
            total_batch = int(dataset.shape[1]/self.batch_size)
            
            for epoch in np.arange(num_epochs):
                for i in np.arange(total_batch):
                    batch_xs =get_next_batch(dataset,self.batch_size)
                    _,c =sess.run([optimizeer,cost],feed_dict={X:batch_xs})
                    if epoch%display_step ==0:
                        print("Epoch:", '%04d' % (epoch+1),"cost=", "{:.9f}".format(c))
            print("Optimization finished!")
        ended =time.time()
        print("Elapsed time:", ended-started)

    def log_func(x1, x2):
        return tf.multiply(x1, tf.log(tf.div(x1,x2)))

    def KL_Div(rho, rho_hat):
        inv_rho = tf.subtract(tf.constant(1.), rho)
        inv_rhohat = tf.subtract(tf.constant(1.), rho_hat)
        log_rho = logfunc(rho,rho_hat) + log_func(inv_rho, inv_rhohat)
        return log_rho

    def load_dataset(self):
        return pd.read_csv("REAL_COMPONENTS.csv")

    def get_next_batch(dataset, batch_size):
        batch= dataset[:batch_size]
        batch_shape = [batch_size] + mrs_data_input_shape
        return np.reshape(batch, batch_shape)
```
